```python
# You may need to install the library first: !pip install scikit-surprise
import pandas as pd
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy

# 1. Load the built-in MovieLens 100k dataset
data = Dataset.load_builtin('ml-100k')

# 2. Split the data into training (75%) and testing (25%) sets
trainset, testset = train_test_split(data, test_size=0.25)

# 3. Use the SVD (Singular Value Decomposition) algorithm
# This is a Matrix Factorization technique used by the famous Netflix Prize winner
model = SVD()

# 4. Train the algorithm on the trainset
model.fit(trainset)

# 5. Make predictions on the testset
predictions = model.test(testset)

# 6. Evaluate the model using RMSE
rmse_score = accuracy.rmse(predictions)
print(f"The Root Mean Squared Error (RMSE) of the model is: {rmse_score:.4f}")

# 7. Predict a specific rating
# Let's predict what user '196' would rate movie '302'
uid = str(196)  # User ID
iid = str(302)  # Item ID
pred = model.predict(uid, iid, verbose=True)
```

```python
# 1. Install Gradio if you haven't already
!pip install -q gradio

import gradio as gr

# 2. Define the Prediction Function
# This uses the 'model' variable from your previous cell
def recommend_movie(user_id, movie_id):
    try:
        # Convert inputs to strings as required by the MovieLens dataset
        uid = str(user_id)
        iid = str(movie_id)

        # Get the prediction from your existing trained model
        prediction = model.predict(uid, iid)

        # Return a user-friendly string
        return f"Predicted Rating: {prediction.est:.2f} / 5.0"
    except Exception as e:
        return f"Error: {str(e)}"

# 3. Create the User Interface
interface = gr.Interface(
    fn=recommend_movie,
    inputs=[
        gr.Textbox(label="User ID", placeholder="Try 196"),
        gr.Textbox(label="Movie ID", placeholder="Try 302")
    ],
    outputs=gr.Text(label="System Recommendation"),
    title="ðŸŽ¬ MovieLens Recommendation System",
    description="Enter a User ID and Movie ID to see the AI's predicted rating based on collaborative filtering."
)

# 4. Launch the public URL
interface.launch(share=True)
```
